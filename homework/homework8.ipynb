```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def make_sin_wave(x, A, w):
    y = A * np.sin(w * x)
    return y
```

```python
x = np.linspace(0, 2*np.pi, 1000)
A_array = np.array([0.5, 1, 1.5, 2, 2.5])
w_array = np.array([1, 2, 3, 4, 5])
```

```python
plt.figure(figsize = (10,5))

for A, w in zip(A_array, w_array):
    y = make_sin_wave(x, A, w)
    plt.plot(x, y, label=f"A={A}, w={w}")

plt.title("Sine Waves with Varying Amplitudes and Frequencies")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
penguins = sns.load_dataset("penguins")
```

```python
penguins
```

```python
body_mass = penguins["body_mass_g"]
bill_length = penguins["bill_length_mm"]
bill_depth = penguins["bill_depth_mm"]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15,5))
sns.scatterplot(x=body_mass, y=bill_length, ax=axs[0], label=("Bill Length"))
sns.scatterplot(x=body_mass, y=bill_depth, ax=axs[0], label=("Bill Depth"))
axs[0].set_title("Bill Size vs Body Mass")
axs[0].set_xlabel("Body Mass (g)")
axs[0].set_ylabel("mm")
axs[0].legend()

sns.histplot(x=body_mass, ax=axs[1], color="purple")
```
